Suppose that they only care about which pizzas are discounted, not
the specific day that each is discounted.<br><br>
For each of the following questions, show working to justify your answer. Evaluate all answers (i.e., do not leave them in terms of combination notation or factorials).<br><br>
How many choices are there if:
</font></p>**

<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">

<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">
<i>Answer:</i><br> 
This is a problem of combination allows repetition, so according to the equation,
$C_{11+6-1}^{6} = \frac{16!}{10!6!}= 8008 $ 
</font></p>

<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">
(b) they want to discount 6 different pizzas?
</font></p>


<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">
<i>Answer:</i><br> 
This is a problem of combination with no repetition, so according to the equation,
$C_{11}^{6}= \frac{11!}{6!5!} = 462 $ 
</font></p>

<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">
(c) they want to discount 6 different pizzas, but the restaurant wants
to sell discounted pizzas containing pepperoni on 2 or 3 nights?
</font></p>


<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">
<i>Answer:</i><br> 
The same as (b), combination with no repetition, but there are 2 cases, one is choose 3 from 8 pizzas(no pepperoni) and all the 3 pepperoni pizzas, the other is choose 4 from 8 pizzas(no pepperoni) and 2 from 3 pepperoni pizzas.
$C_{8}^{3} \times C_{3}^{3} + C_{8}^{4} \times C_{3}^{2} = \frac{8!}{3!5!} \times \frac{3!}{3!0!} + \frac{8!}{4!4!} \times \frac{3!}{2!1!}= 266 $ 
</font></p>

<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">
(d) the same pizza can be discounted on multiple nights, but the
restaurant wants to sell discounted pizzas containing pepperoni
on 2 or 3 nights?
</font></p>

<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">
<i>Answer:</i><br> 
This can be converted to another question, two cases:<br>
    
Case I:<br>
Choose 2 from 3 pepperoni pizzas allows repetition, and choose 4 from 8 pizzas(no pepperoni) allows repetition.<br>
$C_{2+3-1}^{2} \times C_{4+8-1}^{4} =\frac{4!}{2!2!} \times \frac{11!}{4!7!} = 1980$<br><br>
    
Case II:<br>
Choose 3 from 3 pepperoni pizzas allows repetition, and choose 3 from 8 pizzas(no pepperoni) allows repetition.<br>
$C_{3+3-1}^{3} \times C_{3+8-1}^{3} =\frac{5!}{3!2!} \times \frac{10!}{3!7!} = 1200$<br><br>
    
All in all, the total choices are Case I + Case II = 1980 + 1200 = 3180

**<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">(b)$\sum_{n=1}^{\infty}\sqrt[3]{n} $
</font></p>**

<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">
<i>Answer:</i> Divergent, firstly, $\sqrt[3]{n+1} \gt \sqrt[3]{n} \gt 0\ \ (\forall n \gt 0)$, so it is a positive non-decreasing sequence, then, $\lim_{n \to \infty} \sqrt[3]{n}$ doesn't exist because it tends to be $+\infty$, so it can not converge.
</font></p>

**<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">(c)$\sum_{n=1}^{\infty}b_n,\ where\  \sum_{n=1}^{N}b_n = 1+ \frac{1}{N}\ for\ all\ N \ge  1 $
</font></p>**

<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">

## 4(a)
Define the following functions in Python:
### (i) $f(x) = \frac{\displaystyle 1}{\displaystyle 1+x^{2}}$
### (ii) $g(x) = x^{2}-x$

```python
### Enter your answer to question 4.a here

def f(x):
    """
    Calculate the value of the function 1/(1+x**2) for the given input x.

    Args:
        x (float): The input value for the function.

    Returns:
        float: The output value of the function.
    """
    
    result = 1/(1+x**2)
    return result

def g(x):
    """
    Calculate the value of the function (x**2 - x) for the given input x.
    
    Args:
        x (float): The input value for the function.

    Returns:
        float: The output value of the function.
    """
    
    result = x**2 - x
    return result
```

```python
?f
```

## 4(b)
Produce plots for each of the following:
### (i) $f \circ g$
### (ii) $g \circ f$
In each case, you should choose settings (e.g., domain, number of samples) to produce a smooth plot which displays the key features of each function. You may need to experiment with different values to determine what looks good.

<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">(i) <i>Answer: </i> $f \circ g = \frac{\displaystyle 1}{\displaystyle 1+(x^{2}-x)^{2}}$, so the denominator cannot $= 0\ (\forall\ x) $, so domain is $R$.
</font></p>

```python
### Enter your answer to question 4.b.i here

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = f(g(x))

plt.plot(x, y)
plt.xlabel("$x$")
plt.ylabel("$f \circ g$")
plt.title("Graph of $f(g(x))$")
plt.annotate("$D = \{x|x \in R\}$", xy = (-5,0.9))
plt.show()
```

<p style="line-height: 1.5">
<font size = 4px face = "Times New Roman">(ii) <i>Answer: </i> $g \circ f =  \frac{\displaystyle 1}{\displaystyle 1+x^{2}}^{2}-\frac{\displaystyle 1}{\displaystyle 1+x^{2}}$, the same, denominator cannot $= 0\ (\forall\ x) $, so domain is $R$.
</font></p>

```python
### Enter your answer to question 4.b.ii here

x = np.linspace(-5, 5, 100)
y = g(f(x))

plt.plot(x, y)
plt.xlabel("$x$")
plt.ylabel("$g \circ f$")
plt.title("Graph of $g(f(x))$")
plt.annotate("$D = \{x|x \in R\}$", xy = (-5,-0.01))
plt.show()
```
